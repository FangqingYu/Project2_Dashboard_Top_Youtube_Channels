```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
import splinter
import requests
```

```python
url = "https://socialblade.com/youtube/top/100/mostviewed"
response = requests.get(url)
soup = bs(response.text, 'lxml')
```

```python
names = soup.find_all('a', class_="")
channel_names = [name.text for name in names]
```

```python
df = pd.DataFrame(channel_names).rename(columns={0: "name"}).dropna().reset
top100_df = df.loc[121:220,:]
top100_df 
```

```python
df.to_csv("socialblade_top100_byviews.csv")
```
